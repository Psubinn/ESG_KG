```python
import fitz
import pandas as pd

def pdf_pages_to_dataframe(pdf_path, words_list):
    page_data = []
    with fitz.open(pdf_path) as pdf_file:
        for page_num in range(pdf_file.page_count):
            page = pdf_file[page_num]
            text = page.get_text()
            for word in words_list:
                if word.lower() in text.lower():
                    page_data.append({
                        'Page': page_num+1,
                        'Word': word,
                        'Text': text
                    })
                    break
    df = pd.DataFrame(page_data, columns=['Page', 'Word', 'Text'])
    return df

# Example usage
pdf_path = r"C:\Users\user0216\OneDrive - Sejong University\바탕 화면\ESG_classificaion_KG\개인\new\data\(배포용) K-ESG 가이드라인 v1.0.pdf"
words_list = E4
df = pdf_pages_to_dataframe(pdf_path, words_list)
print(df.head())
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def cosine_similarity_dataframe(sentences_list, given_sentence):
    # add given sentence to sentences list
    sentences_list.append(given_sentence)
    
    # vectorize sentences using TfidfVectorizer
    vectorizer = TfidfVectorizer()
    vectors = vectorizer.fit_transform(sentences_list)
    
    # calculate cosine similarity between given sentence and sentences in list
    similarity_scores = cosine_similarity(vectors)[-1][:-1]
    
    # create DataFrame with sentences and their similarity scores
    df = pd.DataFrame({'sentence': sentences_list[:-1], 'similarity Score': similarity_scores})
    
    # sort DataFrame by similarity score in descending order
    df = df.sort_values('similarity Score', ascending=False)
    
    # remove given sentence from list
    sentences_list.pop()
    
    return df

given_sentences = ["에너지 사용량이 당해년도 산업 평균 초과",
                   " 에너지 사용량이 당해년도 산업 평균 동일",
                   "에너지 사용량이 당해년도 산업 평균 미만",
                  ]

first_cos = cosine_similarity_dataframe(output_c, given_sentences[0])
second_cos = cosine_similarity_dataframe(output_c, given_sentences[1])
third_cos = cosine_similarity_dataframe(output_c, given_sentences[2])
#fourth_cos = cosine_similarity_dataframe(output_c, given_sentences[3])
#fifth_cos = cosine_similarity_dataframe(output_c, given_sentences[4])
```

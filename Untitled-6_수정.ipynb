```python
data=bnk[bnk['label']=='E']
data.reset_index(inplace=True)
data.drop(data.columns[0],axis=1, inplace=True)
data
```

```python
""" 키워드 추출 from title """
from konlpy.tag import Kkma
from konlpy.tag import Komoran
from tqdm import tqdm

kkma = Kkma()
komoran = Komoran()

#텍스트 데이터를 리스트로 변환
Data_list=data['text']

#리스트를 요소별로(트윗 하나) 가져와서 명사만 추출한 후 리스트로 저장
data_word=[]
for i in tqdm(range(len(Data_list))):
    try:
        data_word.append(komoran.nouns(Data_list[i]))
    except Exception as e:
        continue

#명사만 추출해서 만든 리스트
data['nouns'] = data_word
data
```

```python
for i in range(len(data)):
    str1 = ' '.join(str(s) for s in data['nouns'][i])
    data['nouns'][i] = str1
data
```

```python
vocab = {}
cnt = 0
for i in data['nouns']:
    i = i.split(' ')
    for j in range(len(i)):
        if i[j] in vocab or len(i[j])<=1:
            cnt+=1
            pass
        else:
            vocab[i[j]] = 0
```

```python
print(len(vocab), vocab)
```

```python
sentences = data['text']

# 찾고자 하는 단어들
words = ['강화',"실천",'구축','명시','실시','제공','지원','기여','참여','동참','방지','대체','인증','수립','이행','실현','인증']

# 각 문장에서 모든 단어를 포함하면 1을, 아니면 0을 반환하는 리스트 생성
result = [1 if any(word in sentence for word in words) else 0 for sentence in sentences]

data['updown'] = result
data
```

```python
len(data[data['updown']==1])
```

```python
up = 123
down = 98
up_ratio = up/(up+down)
down_ratio = down/(up+down)

import collections
for i,w in enumerate(data['nouns']):
    w = w.split(' ')
    if (data.iloc[i]['updown']==1):
        for j in range(len(w)):
            noun = w[j]
            if len(noun)<=1:
              continue
            vocab[noun] = vocab[noun] + down_ratio
    else:
        for j in range(len(w)):
            noun = w[j]
            if len(noun)<=1:
              continue
            vocab[noun] = vocab[noun] - up_ratio
```

```python
vocab
```

```python
total = []
for i,w  in enumerate(data['nouns']):
    sent_score = 0
    w= w.split(' ')
    for j in w:
        if(len(j)<=1):
          continue
        elif(j not in vocab):
          continue
        else:
          sent_score = sent_score + vocab[j]
    total.append(sent_score/len(w))
data['sent_score'] = total
```

```python
sum = 0
for i in range(len(vocab)):
  sum = sum + list(vocab.values())[i]
sent_mean = sum/len(vocab)

a_ = []
for i in range(len(data)):
  if(data.iloc[i]['sent_score']>sent_mean):
    a_.append(1)
  else:
    a_.append(0)
data['sent_label'] = a_
data
```

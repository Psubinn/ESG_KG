```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 예시 문장 데이터
sentences = data['text']

# 예시 단어 데이터
keyword = words[6]

# TfidfVectorizer로 문장과 단어 벡터화
tfidf_vectorizer = TfidfVectorizer(use_idf=True)
sentences_vectors = tfidf_vectorizer.fit_transform(sentences)
keyword_vector = tfidf_vectorizer.transform([keyword])

# 문장과 단어 벡터간 코사인 유사도 계산
cosine_similarities = cosine_similarity(sentences_vectors, keyword_vector)

# 결과를 DataFrame으로 출력
df = pd.DataFrame(cosine_similarities, columns=['similarity'])
df['text'] = ''
df['text'] = sentences
df
```

```python
df.sort_values(by='similarity',ascending=False,inplace=True)
df.reset_index(inplace=True)
df.drop(df.columns[0],axis=1,inplace=True)
df
```

```python
def vader_polarity(review, threshold = 0.05):
    from nltk.sentiment.vader import SentimentIntensityAnalyzer
    
    # VADER 객체로 감성 지수 산출
    analyzer = SentimentIntensityAnalyzer()
    scores = analyzer.polarity_scores(review)
    
    # 감성 지수가 threshold 보다 크거나 같으면 1, 그렇지 않으면 0
    agg_score = scores['compound']
    final_sentiment = 1 if agg_score >= threshold else 0
        
    return final_sentiment
```

```python
# 각 문서별로 긍정/부정 예측
bnk_1['vader_preds'] = bnk_1['text'].apply( lambda x : vader_polarity(x, 0.1) )

y_target = bnk_1['sentiment'].values
vader_preds = bnk_1['vader_preds'].values
```

```python
def sentiment_analysis(text):
    sid = SentimentIntensityAnalyzer()
    score = sid.polarity_scores(text)
    if score['compound'] >= 0.05:
        return 'positive'
    elif score['compound'] <= -0.05:
        return 'negative'
    else:
        return 'neutral'
```

```python
from tqdm import tqdm

a=[]
b=[]
sentences = df['text']
for sentence in tqdm(sentences):
    sentiment = sentiment_analysis(sentence)
    if sentiment is not 'neutral':
        a.append(sentence)
        b.append(sentiment)
```

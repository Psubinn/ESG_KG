```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 주어진 DataFrame
df = bnk

# 키워드 리스트
keywords = ['환경경영 목표', '재생가능한 원자재 및 재활용 또는 포장용 재료', '온실가스', '에너지', '용수 및 폐수', '폐기물', '오염 물질']

# TfidfVectorizer를 사용하여 문장 벡터화
vectorizer = TfidfVectorizer()
text_vectors = vectorizer.fit_transform(df['text'])

# 개별 키워드별 DataFrame 생성
keyword_dfs = {}
for i, keyword in enumerate(keywords):
    # 유사도 계산
    keyword_vector = vectorizer.transform([keyword])
    similarity_matrix = cosine_similarity(text_vectors, keyword_vector)
    similarities = similarity_matrix.flatten()

    # DataFrame 생성
    keyword_df = pd.DataFrame({'발행년도': df['발행년도'],
                               'label': df['label'],
                               'text': df['text'],
                               '유사도': similarities})
    keyword_df['keyword'] = keyword

    # 유사도가 큰 순서대로 정렬하여 저장
    keyword_df = keyword_df.sort_values(by='유사도', ascending=False)
    keyword_dfs[f"bnk_{i+1}"] = keyword_df

# 결과 저장
for name, keyword_df in keyword_dfs.items():
    globals()[name] = keyword_df
```

```python
combined_df = pd.concat(keyword_dfs.values())
combined_df
```

```python
# 'similarity'가 0인 행 제거
combined_df = combined_df[combined_df['유사도'] == 0]

# 결과 출력
combined_df.reset_index(inplace=True)
combined_df.drop(combined_df.columns[0],axis=1,inplace=True)
combined_df

combined_df.to_csv('negative.csv')
```

```python
import pandas as pd
from transformers import BertTokenizer, BertForSequenceClassification
import torch

# 미리 학습된 BERT 모델과 토크나이저 로드
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased', do_lower_case=False)
model = BertForSequenceClassification.from_pretrained("bert-base-multilingual-cased", num_labels=7)

# 입력 문장들과 측면(Aspect) 정의
sentences = full['text']

aspects = ["에너지", "오염물질", "온실가스", "용수","폐기물","환경경영 목표","재생가능한 원자재 및 재활용 또는 포장용 재료"]

# 문장별 측면별 감성 점수 계산
results = []
for i, sentence in tqdm(enumerate(sentences)):
    scores = []
    for aspect in (aspects):
        input_text = f"[CLS] {aspect}: {sentence} [SEP]"
        inputs = tokenizer(input_text, truncation=True, padding=True, return_tensors="pt")
        outputs = model(**inputs)
        logits = outputs.logits
        score = torch.softmax(logits, dim=1).tolist()[0][1]
        scores.append(score)
    results.append(scores)

# 결과를 데이터프레임 형태로 출력
df = pd.DataFrame(results, columns=aspects)
df['기업'] = full['기업']
df = df[['기업'] + aspects]  # '기업' 열을 첫 번째 열로 위치 조정

# 결과 출력
df
```

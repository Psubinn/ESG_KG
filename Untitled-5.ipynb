```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 주어진 DataFrame
df = bnk

# 키워드 리스트
keywords = ['환경경영 목표', '재생가능한 원자재 및 재활용 또는 포장용 재료', '온실가스', '에너지', '용수 및 폐수', '폐기물', '오염 물질']

# TfidfVectorizer를 사용하여 문장 벡터화
vectorizer = TfidfVectorizer()
text_vectors = vectorizer.fit_transform(df['text'])

# 개별 키워드별 DataFrame 생성
keyword_dfs = {}
for i, keyword in enumerate(keywords):
    # 유사도 계산
    keyword_vector = vectorizer.transform([keyword])
    similarity_matrix = cosine_similarity(text_vectors, keyword_vector)
    similarities = similarity_matrix.flatten()

    # DataFrame 생성
    keyword_df = pd.DataFrame({'발행년도': df['발행년도'],
                               'label': df['label'],
                               'text': df['text'],
                               '유사도': similarities})
    keyword_df['keyword'] = keyword

    # 유사도가 큰 순서대로 정렬하여 저장
    keyword_df = keyword_df.sort_values(by='유사도', ascending=False)
    keyword_dfs[f"bnk_{i+1}"] = keyword_df

# 결과 저장
for name, keyword_df in keyword_dfs.items():
    globals()[name] = keyword_df
```

```python
def vader_polarity(review, threshold = 0.05):
    from nltk.sentiment.vader import SentimentIntensityAnalyzer
    
    # VADER 객체로 감성 지수 산출
    analyzer = SentimentIntensityAnalyzer()
    scores = analyzer.polarity_scores(review)
    
    # 감성 지수가 threshold 보다 크거나 같으면 1, 그렇지 않으면 0
    agg_score = scores['compound']
    final_sentiment = 1 if agg_score >= threshold else 0
        
    return final_sentiment
```

```python
# 각 문서별로 긍정/부정 예측
bnk_1['vader_preds'] = bnk_1['text'].apply( lambda x : vader_polarity(x, 0.1) )

y_target = bnk_1['sentiment'].values
vader_preds = bnk_1['vader_preds'].values
```

```python
def sentiment_analysis(text):
    sid = SentimentIntensityAnalyzer()
    score = sid.polarity_scores(text)
    if score['compound'] >= 0.05:
        return 'positive'
    elif score['compound'] <= -0.05:
        return 'negative'
    else:
        return 'neutral'
```

```python
from tqdm import tqdm

a=[]
b=[]
sentences = df['text']
for sentence in tqdm(sentences):
    sentiment = sentiment_analysis(sentence)
    if sentiment is not 'neutral':
        a.append(sentence)
        b.append(sentiment)
```

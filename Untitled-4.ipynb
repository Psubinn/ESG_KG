```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 예시 문장 데이터
sentences = data_e['text']

# 예시 단어 데이터
keyword = words[0]

# TfidfVectorizer로 문장과 단어 벡터화
tfidf_vectorizer = TfidfVectorizer(use_idf=True)
sentences_vectors = tfidf_vectorizer.fit_transform(sentences)
keyword_vector = tfidf_vectorizer.transform([keyword])

# 문장과 단어 벡터간 코사인 유사도 계산
cosine_similarities = cosine_similarity(sentences_vectors, keyword_vector)

# 결과를 DataFrame으로 출력
df = pd.DataFrame(cosine_similarities, columns=['similarity'])
df['text'] = ''
df['text'] = sentences
df
```

```python
df.sort_values(by='similarity',ascending=False,inplace=True)
df.reset_index(inplace=True)
df.drop(df.columns[0],axis=1,inplace=True)
df
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
def sentiment_analysis(text):
    sid = SentimentIntensityAnalyzer()
    score = sid.polarity_scores(text)
    if score['compound'] >= 0.03:
        return 'positive'
    elif score['compound'] <= -0.03:
        return 'negative'
    else:
        return 'neutral'
```

```python
from tqdm import tqdm

a=[]
b=[]
sentences = e2['text']
for sentence in tqdm(sentences):
    sentiment = sentiment_analysis(sentence)
    if sentiment is not 'neutral':
        a.append(sentence)
        b.append(sentiment)
```
